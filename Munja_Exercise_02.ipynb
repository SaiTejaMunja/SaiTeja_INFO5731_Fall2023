Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
# You code here (Please add comments in the code):
import requests
import pandas as pd
from bs4 import BeautifulSoup

imdb_url = "https://www.imdb.com/search/title?count=100&start={}&title_type=feature,tv_series&ref_=nv_wl_img_2"
rt_url = "https://www.rottentomatoes.com/m/{}"

def get_data_imbd(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    hollywood_movies = soup.select("h3 > a")
    list1 = []
    for movie in hollywood_movies:
        title_of_movie = movie.get_text()
        id_of_movie = movie.attrs["href"].split("/")[2]
        list1.append([title_of_movie, id_of_movie])
    return list1

def get_rt_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    try:
        movie_rating = soup.select_one(".mop-ratings-wrap__percentage").get_text()
    except:
        movie_rating = None
    return movie_rating

list2 = []
for i in range(10):
    count = i * 100 + 1
    data_imbd = get_data_imbd(imdb_url.format(count))
    for movie in data_imbd:
        title_of_movie, id_of_movie = movie
        rt_data = get_rt_data(rt_url.format(id_of_movie))
        list2.append([title_of_movie, rt_data])

data_frame = pd.DataFrame(list2, columns=["Title", "Rotten Tomatoes Rating"])


# Write to CSV File
data_frame.to_csv("rotten_tomatos.csv", index=False)

data_frame
```

Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2013-2023).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# You code here (Please add comments in the code):

import requests
from bs4 import BeautifulSoup
import csv

# Define the base URL and search query
base_url = "https://scholar.google.com"
search_query = "/scholar?q=information+retrieval&as_ylo=2013&as_yhi=2023"

# Define headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Function to extract information from a search page
def extract_info_from_page(url):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    articles = soup.find_all("div", class_="gs_ri")

    article_info = []

    for article in articles:
        title = article.find("h3").text
        authors = article.find("div", class_="gs_a").text
        venue = article.find("div", class_="gs_a").text
        year = article.find("div", class_="gs_a").text
        abstract = article.find("div", class_="gs_rs").text

        article_info.append((title, venue, year, authors, abstract))

    return article_info

# Function to save the information to a CSV file
def save_to_csv(data, filename):
    with open(filename, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["Title", "Venue", "Year", "Authors", "Abstract"])
        writer.writerows(data)

# Main function
def main():
    all_articles = []

    for page in range(0, 91, 10):  # Iterate through search result pages
        url = f"{base_url}{search_query}&start={page}"
        articles = extract_info_from_page(url)
        all_articles.extend(articles)

        if len(all_articles) >= 1000:
            break

    save_to_csv(all_articles[:1000], "articles_info.csv")

if __name__ == "__main__":
    main()
```
